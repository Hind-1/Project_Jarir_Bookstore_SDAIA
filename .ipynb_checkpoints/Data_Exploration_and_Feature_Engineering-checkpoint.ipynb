```python
df=pd.read_csv('/Users/hindsalem/Desktop/tempjjh/Sheet 2-Table 1.csv')
```

```python
df.Ram=df.Ram.str.replace('GB','').astype(float)
df.Weight=df.Weight.str.replace('kg','').astype(float)
```

```python
df.info()
```

```python
df['IPS']=df['ScreenResolution'].apply(lambda ips: 1 if 'IPS' in ips else 0)
```

```python
df['Touchscreen']=df.ScreenResolution.apply(lambda val_of_serise: 1 if 'Touchscreen'in val_of_serise else 0)

```

```python
df
```

```python
Accuracy=df['ScreenResolution'].str.split('x',expand=True)
# Accuracy
df['X']=Accuracy[0]

df['Y']=Accuracy[1]
```

```python
len(df['X'])
```

```python
st=df.X.str.findall(r'(\d+\.?\d+)').apply(lambda x: x[0])
df.X=st
```

```python
df
```

```python
df.X=df.X.astype(int)
df.Y=df.Y.astype(int)
```

```python
df.info()
```

```python
df.corr()['Price']
```

```python
df['ppi']=((((df.X)**2)+((df.Y)**2)**0.5)/(df.Inches))
```

```python
df.corr()['Price']
```

```python
df.drop(columns=['Inches','X','Y','ScreenResolution'],inplace=True)
df
```

```python
 df['Cpu']= df['Cpu'].apply(lambda x:" ".join(x.split()[0:3]))
```

```python
def processor(text):
    if text == 'Intel Core i7' or text == 'Intel Core i5' or text == 'Intel Core i3':
        return text
    else:
        if text.split()[0] == 'Intel':
            return 'Other Intel Processor'
        else:
            return 'AMD Processor'
```

```python
df['Cpu'] = df['Cpu'].apply(processor)
```
